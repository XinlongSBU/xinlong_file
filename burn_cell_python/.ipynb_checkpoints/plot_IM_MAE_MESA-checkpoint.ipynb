```python
import yt
import numpy as np
import argparse
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# Read data from mesa data: 
a = pd.read_excel("/Users/sailor/Desktop/asro_work/git_fun/xinlong_file/zingale_mesa/ONe6040_update.xlsx")

# read data from Maestro data:
ds = yt.load("ECSN_full/plt0000001/")
name = ds.fullpath.split('/')[-1]

# read data form initial model
f = open("/Users/sailor/Desktop/asro_work/git_fun/xinlong_file/zingale_mesa/kepler_xinlong_ONe6040-final.hse.5120","r")
data = f.readlines()  # data is a list. each element is a line of "23Ne...dat"
f.close()
del data[0:16]

data2d = []
for i in range(len(data)):
    data2d.append(re.split(r"[ ]",data[i].strip('\n')))

for i in range(len(data2d)):
        while '' in data2d[i]:
            data2d[i].remove('')

for i in range(len(data2d)-1,-1,-1):
    if len(data2d[i]) == 3:
        data2d.pop(i)
        #ll.append(i)
        i+=1
    else:
        data2d[i].append('0.9999999989999999E-10')
        data2d[i].append('0.9999999989999999E-10')
        data2d[i].append('0.9999999989999999E-10')
        #data2d[i].append('0.9999999989999999E-10')
        #mm.append(i)

data3d = np.array(data2d).astype(np.float)

df = pd.DataFrame(data3d)
df.columns = ['radius','density','temperature','pressure','ne20','o16','o20','mg24','f29','he4','al27','si28','s32','p31','h1']
df.head()
```

```python
# Create a sphere of radius 1500 km in the center of the box.
my_sphere = ds.sphere("c", (1500.0, "km"))
```

```python
plot = yt.ProfilePlot(my_sphere, "radius", "Hnuc", weight_field="cell_volume")
plot.set_log('radius', False)
plot.set_xlim(0, 1.5e8)
profile = plot.profiles[0]

to_save = pd.DataFrame(columns = ['radius'], data = profile.x)
to_save['Hnuc [erg/g/s]'] = profile['Hnuc']

r_sun = 6.955e10  # cm
radius = np.array(a["radius"])
eps_nuc2 = a['eps_nuc_mc2']
eps_nuc_neu = a['eps_nuc_neu']
#e_nuc = eps_nuc2-eps_nuc_neu
e_nuc = eps_nuc_neu

plt.semilogy(radius*r_sun,e_nuc, label = 'MESA')
plt.semilogy(profile.x, profile['Hnuc'], label = 'plt0000000')
plt.xlabel("radius [cm]")
plt.ylabel("e_nuc [erg/g/s]")
plt.ylim(1e5,1e13)
plt.xlim(0,1.5e8)
plt.grid()
plt.legend()
plt.savefig(name+"_enuc.png")
```

```python
plot = yt.ProfilePlot(my_sphere, "radius", "h", weight_field="cell_volume")
plot.set_log('radius', False)
plot.set_xlim(0, 1.5e8)
profile = plot.profiles[0]

to_save['h [erg/g]'] = profile['h']

plt.plot(profile.x, profile['h'], label = 'plt0000000')
plt.xlabel("radius [cm]")
plt.ylabel("enthalpy [erg/g]")
plt.xlim(0,1.5e8)
plt.grid()
plt.legend()
plt.savefig(name+"_enthalpy.png")
```

```python
plot = yt.ProfilePlot(my_sphere, "radius", "tfromp", weight_field="cell_volume")
plot.set_log('radius', False)
plot.set_xlim(0, 1.5e8)
profile = plot.profiles[0]

to_save['tfromp'] = profile['tfromp']

r_sun = 6.955e10  # cm
radius = np.array(a["radius"])
temp = np.array(a["temperature"])

plt.plot(radius*r_sun, temp, label = 'MESA')
plt.plot(profile.x, profile['tfromp'], label = 'MAESTROeX')
plt.plot(df['radius'],df['temperature'],'--', label = 'initial model')
plt.xlabel("radius [cm]")
plt.ylabel("temperature [K]")
#plt.ylim(1e6,1e13)
plt.xlim(0,1.5e8)
plt.grid()
plt.legend()
plt.savefig(name+"_temperature.png")
```

```python
plot = yt.ProfilePlot(my_sphere, "radius", "p0", weight_field="cell_volume")
plot.set_log('radius', False)
plot.set_xlim(0, 1.5e8)
profile = plot.profiles[0]

to_save['p0'] = profile['p0']

r_sun = 6.955e10  # cm
radius = np.array(a["radius"])
temp = np.array(a["pressure"])

plt.plot(radius*r_sun, temp, label = 'MESA')
plt.plot(profile.x, profile['p0'], label = 'MAESTROeX')
plt.plot(df['radius'],df['pressure'],'--', label = 'initial model')
plt.xlabel("radius [cm]")
plt.ylabel("pressure [Pa]")
#plt.ylim(1e6,1e13)
plt.xlim(0,1.5e8)
plt.grid()
plt.legend()
plt.savefig(name+"_pressure.png")
```

```python
plot = yt.ProfilePlot(my_sphere, "radius", "rho", weight_field="cell_volume")
plot.set_log('radius', False)
plot.set_xlim(0, 1.5e8)
profile = plot.profiles[0]

to_save['rho'] = profile['rho']

r_sun = 6.955e10  # cm
radius = np.array(a["radius"])
rho = np.power(10,np.array(a["logRho"]))

plt.plot(radius*r_sun, rho, label = 'MESA')
plt.plot(profile.x, profile['rho'],label = 'MAESTROeX')
plt.plot(df['radius'],df['density'],'--', label = 'initial model')
plt.xlabel("radius [cm]")
plt.ylabel("density [g/cm3]")
#plt.ylim(1e6,1e13)
plt.xlim(0,1.5e8)
plt.grid()
plt.legend()
plt.savefig(name+"_density.png")
```
